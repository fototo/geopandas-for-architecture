```python
import geopandas as gpd
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt
from shapely.geometry import LineString
import os

import warnings
warnings.filterwarnings('ignore')
```

```python
gdb_path = os.path.expanduser('~/work/test-data/campus.gdb')
buildings = gpd.read_file(gdb_path, layer=0)  
print(buildings.head())
```

```python
rooms_all = gpd.read_file(gdb_path, layer=1)

# 1005037 3F for bsrb
rooms = rooms_all[(rooms_all["bldrecnbr"]=="1005037") & (rooms_all["FLOOR"] == '03')]
print(rooms.head())
```

```python
# graphics layer and filter it with building and floor
graphics_all = gpd.read_file(gdb_path, layer=4)
graphics = graphics_all[(graphics_all["BldRecNbr"]=="1005037") & (graphics_all["Floor"] == '03')]
```

```python
print(graphics.head())
```

```python
# then plot the nodes over room polygons
fig, ax = plt.subplots()
fig.set_size_inches(6, 6)
ax.axison = False
rooms.plot(ax=ax, linewidth=0.3, color='#DDDDDD')
graphics.plot(ax=ax, markersize=2, color='#6677AA')
ax.set_aspect('equal')
plt.show()
```
